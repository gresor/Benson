```python
def Turnstile_dict(csvfile):
    """ takes a csv file with turnsile data, loads it into a dictionary making a key for each turnstile"""
    with open(csvfile,'r') as turnstile_file:
        turn ={}
        #turn is the dictionary
        reader = csv.reader(turnstile_file)
        for num, row in enumerate(reader):
            if num >0: #this gets rid of the header row
                k=tuple(row[0:4]) #makes the key that corrisponds to the turnstile
                if k not in turn.keys(): 
                    turn[k]=[row[4:11]] # put the key and the keys first list in the dict if it didn't exist
                else:
                    turn[k].append(row[4:11]) #appends lists when the key arleady exists
    return turn    

       # this make a dectionary where the turnstiles are keys       
```

```python
def tuple_to_two_list(lt):
    """takes a list of tuples and makes it two different lists"""
    list1=[]
    list2=[]
    for i in range(len(lt)):
        list1.append(lt[i][0])
        list2.append(lt[i][1])
    return list1, list2
```

```python
def Station_Sum(d, station_sum =None):
    """sums all the values for staion, pass in a dictionary d of station counts by day
    optionally pas in another dictiony with station sums from other weeks"""
    if station_sum is None:
        station_sum={}
    for sk in d.keys():
        if sk not in station_sum.keys():
            tot=0
            for (day, count) in d[sk]:
                tot+=count
            station_sum[sk]=tot
        else:
            for (day,count) in d[sk]:
                station_sum[sk]+=count
    return station_sum
        
```
